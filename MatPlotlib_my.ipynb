```python
fig = plt.figure() #blank canvas

axes = fig.add_axes([0.1,0.1,0.8,0.8])
#list with left bottom width and height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=3, ncols=3)

#axes.plot(x,y)
plt.tight_layout()  #fixing overlaps
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)  #does the job for add_axes

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)  #does the job for add_axes

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'r',label='X Squared')
ax.plot(x,x**3,'g',label='X Cubed')

ax.legend(loc=0)  #avoiding overlap with legend box '=0' automatically checks for the best one
```
